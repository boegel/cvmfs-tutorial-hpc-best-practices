```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from seaborn import boxplot, violinplot
```

```python
data_cold_cache_local_cold = {
    "RAM disk": dev_shm_cold_cache,
    "local SSD": tmp_cold_cache,
}
data_cold_cache_local_hot = {
     "RAM disk": dev_shm_hot_cache,
    "local SSD": tmp_hot_cache,   
}
fontsize = 14
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
fig.suptitle("TensorFlow start-up performance: local disk vs RAM disk", fontsize=fontsize)

violinplot(data_cold_cache_local_cold, inner="point", ax=axes[0])
axes[0].tick_params(axis='both', labelsize=fontsize)
axes[0].set_title("")
axes[0].set_xlabel("cold kernel file system cache", style='italic', fontsize=fontsize)
axes[0].set_ylabel("time (seconds)", fontsize=fontsize)
y_values = [0,1,2,3,4,5]
axes[0].set_ylim((1, max(y_values)))
for y in y_values:
    axes[0].axhline(y=y, color='lightgray', linestyle='--', linewidth=1, zorder=0)
axes[0].set_ylim(0)

violinplot(data_cold_cache_local_hot, inner="point", ax=axes[1])
axes[1].tick_params(axis='both', labelsize=fontsize)
axes[1].set_title("")
axes[1].set_xlabel("hot kernel file system cache", style='italic', fontsize=fontsize)
axes[1].set_ylabel("time (seconds)", fontsize=fontsize)
y_values = [0,1,2,3,4,5]
axes[1].set_ylim((1, max(y_values)))
for y in y_values:
    axes[1].axhline(y=y, color='lightgray', linestyle='--', linewidth=1, zorder=0)
axes[1].set_ylim(0)
```
